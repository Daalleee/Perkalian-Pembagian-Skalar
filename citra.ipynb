```python
import cv2
from matplotlib import pyplot as plt

# Baca gambar
img = cv2.imread('/content/daun.jpg')

# ubah BGR → RGB agar warna benar di matplotlib
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Perkalian skalar
img_kali = cv2.multiply(img, 3)

# Pembagian skalar
img_bagi = cv2.divide(img, 3)

# Plot hasil dalam grid
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.imshow(img)
plt.title("Asli")
plt.axis("off")

plt.subplot(1, 3, 2)
plt.imshow(img_kali)
plt.title("Perkalian x2")
plt.axis("off")

plt.subplot(1, 3, 3)
plt.imshow(img_bagi)
plt.title("Pembagian /2")
plt.axis("off")

plt.tight_layout()
plt.show()
```

```python
import cv2
from matplotlib import pyplot as plt

# Baca gambar
img = cv2.imread('/content/daun.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Perkalian skalar
img_kali = cv2.multiply(img, 3)

# Pembagian skalar
img_bagi = cv2.divide(img, 3)

# Contoh ambil nilai piksel pada koordinat (100, 50)
# Format: [row, col] → [y, x]
y, x = 200, 100

print("Pixel asli:", img[y, x])       # [R, G, B] asli
print("Pixel dikali 3:", img_kali[y, x])
print("Pixel dibagi 3:", img_bagi[y, x])
```
